```python
import numpy as np
np.seterr(divide='ignore', invalid='ignore')

def cromaticCoords(img):
      R = img[:,:,0]
      G = img[:,:,1]
      B = img[:,:,2]
    
      new_img = np.zeros(np.shape(img))
      r  = R/(R+G+B)
      g  = G/(R+G+B)
      b  = B/(R+G+B)
    
      new_img[:,:,0] = r
      new_img[:,:,1] = g
      new_img[:,:,2] = b
        
      return new_img
```
